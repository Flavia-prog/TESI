```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

from aijack.collaborative import FedAVGClient, FedAVGServer, FedAVGAPI

# tiny dummy model
class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc = nn.Linear(10, 2)

    def forward(self, x):
        return self.fc(x)

# dummy data
X = torch.randn(20, 10)
y = torch.randint(0, 2, (20,))
dataset = TensorDataset(X, y)
dataloader = DataLoader(dataset, batch_size=5)

clients = [
    FedAVGClient(SimpleNN()),
    FedAVGClient(SimpleNN())
]

server = FedAVGServer(clients, SimpleNN())

criterion = nn.CrossEntropyLoss()
optimizers = [torch.optim.SGD(c.parameters(), lr=0.1) for c in clients]
dataloaders = [dataloader, dataloader]

api = FedAVGAPI(
    server,
    clients,
    criterion,
    optimizers,
    dataloaders,
    num_communication=1
)

api.run()

print("FedAvg ran successfully")
```
